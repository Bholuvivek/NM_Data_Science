```python
#Read the csv file
DataFrame = pd.read_csv('/content/fake_news.csv')
```

```python
data = DataFrame
```

```python
data.Label.value_counts()
```

```python
#find the null values
data.isnull().sum()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data.Label=le.fit_transform(data.Label)
```

```python
data.head()
```

```python
import nltk
nltk.download('all')
```

```python
from nltk.stem import WordNetLemmatizer
le = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
```

```python
import re
```

```python
datas = []
for i in range(0,len(data)):
  review = re.sub('[^a-zA-Z]'," ",data["Statement"][i])
  review = review.lower()
  review = review.split()
  review = [le.lemmatize(word) for word in review if not word in stopwords.words("english")]
  review = " ".join(review)
  datas.append(review)
```

```python
 datas
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv =TfidfVectorizer(max_features=2500)
X = tv.fit_transform(datas).toarray()
Y = data.Label
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0,stratify=Y)
```

```python
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
lr_ans = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score1 = accuracy_score(Y_test,lr_ans)
```

```python
print(score1)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, Y_train)
```

```python
svc_ans = svc.predict(X_test)
```

```python
score2 = accuracy_score(Y_test, svc_ans)
print(score2)
```

```python
input_news = ["Every poll said I won the debate."]
input_news = tv.transform(input_news)
prediction = lr.predict(input_news)
print(prediction)
if(prediction[0]== 1):
  print("This is True news")
else:
  print("This is Fake news")
```
