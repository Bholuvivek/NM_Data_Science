```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

# **Question 1**
**What is the overall survival rate of passengers on the Titanic**

```python
survival=pd.Series(titanic['alive'].value_counts())
print(survival,"\n")
```

# **Visualizing the filtered data**

```python

plt.pie(survival,labels=['dead','alive'],autopct="%.2f")
plt.legend()
plt.title('Survival % of Passengers')
plt.show()
```

```python
titanic.head()
```

# **Question 2 **
**What is the gender distribution among the passengers on the Titanic**

```python
gender_dist=pd.Series(titanic['sex'].value_counts())
print(gender_dist,"\n")
```

# Visualizing the filtered data

```python

plt.title("Gender Distribution")
plt.xlabel("sex")
plt.ylabel("no. of passengers")
sns.barplot(x=gender_dist.index, y=gender_dist.values, data=titanic)
```

```python
titanic.head()
```

# **#Question 3**
**Did the survival rate differ by gender? If so, by how much?**

```python
gender_surv=titanic.groupby('sex')['alive'].value_counts()
print(gender_surv,'\n')
```

# **Visualizing the filtered data**

```python
plt.title("Gender Distribution Based on Survival")
sns.barplot(x='sex', y='survived', data=titanic)
```

```python
titanic.head()
```

#Question 4
**What was the age distribution among the passengers on the Titanic?**

```python
age_dist=pd.Series(titanic['age'].value_counts().sort_values(ascending=False))
print(age_dist,'\n')
```

# **Visualizing the filtered data**

```python

plt.hist(titanic['age'],bins=10,linewidth=1.5,linestyle='-',edgecolor='white',color='green')
plt.title('Age Dsitribution')
plt.xlabel('age')
plt.ylabel('no. of Passengers')
plt.show()
```

```python
titanic.head()
```

# **Question 5**
 **Did the survival rate differ by the journey class? If so, by how much?**

```python
pclass_surv=titanic.groupby('pclass')['alive'].value_counts()
print(pclass_surv,'\n')
```

# **Visualizing the filtered data**

```python
plt.title("Survival Rate Based on Journey Class")
sns.barplot(x='pclass', y='survived', data=titanic)
```
